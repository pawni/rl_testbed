```python
mc = MountainCar()
print mc.getValidActions()
```

```python
from keras.models import Sequential
from keras.layers.core import Dense, Dropout, Activation

network = Sequential()
network.add(Dense(200, init='lecun_uniform', input_shape=(2,)))
network.add(Activation('tanh'))
network.add(Dense(200, init='lecun_uniform', input_shape=(200,)))
network.add(Activation('tanh'))
network.add(Dense(3, init='lecun_uniform', input_shape=(200,)))
network.add(Activation('softmax'))
network.compile(loss='mse', optimizer='sgd')

eg = eGreedy(0, mc.getNumActions())

mem = Memory(80, 40)

ag = DQN(network, policy=eg, memory = None)
```

```python
ag.evaluate(mc)
```

```python
rews = ag.train(mc, 3000)
```

```python
print rews
```
